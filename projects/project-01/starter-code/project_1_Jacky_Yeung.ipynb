##### 13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python
def standard_deviation_list(list_data):
        
    return (sum([ (d-(sum(list_data)/len(list_data)))**2 for d in list_data])/len(list_data))**0.5
   
for col in SAT_value_dict:
    print col + " standard deviation = " + str(standard_deviation_list(SAT_value_dict[col]))
    
```

##### 19. Plot some scatterplots. **BONUS**: Use a PyPlot `figure` to present multiple plots at once.

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()

ax1.plot(SAT_value_dict['Rate'],SAT_value_dict['Math'], 'b.')
ax1.set_xlabel('Rate')
# Make the y-axis label and tick labels match the line color.
ax1.set_ylabel('Math', color='b')
ax1.set_yticks(range(400,700,50))
for tl in ax1.get_yticklabels():
    tl.set_color('b')


ax2 = ax1.twinx()
ax2.plot(SAT_value_dict['Rate'],SAT_value_dict['Verbal'], 'r.')
ax2.set_ylabel('Verbal', color='r')
ax2.set_yticks(range(400,700,50))
for tl in ax2.get_yticklabels():
    tl.set_color('r')
plt.title("Scatter Math and Verbal vs Rate")
plt.show()
```
