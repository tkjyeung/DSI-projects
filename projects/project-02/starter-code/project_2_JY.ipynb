##### Do some rudimentary cleaning. Rename any columns that are poorly named, shorten any strings that may be too long, fill in any missing values. Explain your rationale for the way you choose to "impute" the missing data.

```python
#no need year column, all are 2000
df_clean = df.ix[:,1:] 
```

```python
#better titles
df_clean = df_clean.rename(columns={'artist.inverted': 'artist', 'date.entered':'date_entered','date.peaked':'date_peaked'})
```

```python
# function to convert time to seconds for easy comparison
def time_to_sec(s):
    return int(s.split(':')[0])*60 + int(s.split(':')[1])
```

```python
df_clean.time = df_clean.time.apply(time_to_sec)
```

```python
#change the string date to datetime object df.day + df.month + df.year, format="%d%m%Y"
df_clean.date_entered = pd.to_datetime(df_clean.date_entered, format='%Y-%m-%d' )
df_clean.date_peaked = pd.to_datetime(df_clean.date_peaked, format='%Y-%m-%d')
```

```python
the_billboard_sorted = []
for week_billboard in the_billboard :
    week_rank = []
    #track_rank = []
    for k in week_billboard :
        week_rank.append(int(k))
    week_rank.sort()
    the_billboard_sorted.append([(wr, week_billboard[wr]) for wr in week_rank])
       
```
