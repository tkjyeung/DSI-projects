VARIABLE DESCRIPTIONS:
survival        Survival
                (0 = No; 1 = Yes)
pclass          Passenger Class
                (1 = 1st; 2 = 2nd; 3 = 3rd)
name            Name
sex             Sex
age             Age
sibsp           Number of Siblings/Spouses Aboard
parch           Number of Parents/Children Aboard
ticket          Ticket Number
fare            Passenger Fare
cabin           Cabin
embarked        Port of Embarkation
                (C = Cherbourg; Q = Queenstown; S = Southampton)

SPECIAL NOTES:
Pclass is a proxy for socio-economic status (SES)
 1st ~ Upper; 2nd ~ Middle; 3rd ~ Lower

Age is in Years; Fractional if Age less than One (1)
 If the Age is Estimated, it is in the form xx.5

With respect to the family relation variables (i.e. sibsp and parch)
some relations were ignored.  The following are the definitions used
for sibsp and parch.

Sibling:  Brother, Sister, Stepbrother, or Stepsister of Passenger Aboard Titanic
Spouse:   Husband or Wife of Passenger Aboard Titanic (Mistresses and Fiances Ignored)
Parent:   Mother or Father of Passenger Aboard Titanic
Child:    Son, Daughter, Stepson, or Stepdaughter of Passenger Aboard Titanic

Other family relatives excluded from this study include cousins,
nephews/nieces, aunts/uncles, and in-laws.  Some children travelled
only with a nanny, therefore parch=0 for them.  As well, some
travelled with very close friends or neighbors in a village, however,
the definitions do not support such relations.

# Passenger Class

```python
titanic_df.Pclass.value_counts()
```

```python
x1 = titanic_df.Pclass.value_counts().plot.bar(title = "Passenger Class Distribution")
x1.set_xlabel("Passenger Class")
x1.set_ylabel("Passenger Count")
```

```python
groups = titanic_df.groupby(by="Survived")
df1 = groups.Pclass.value_counts()
df1
```

```python
x2 = groups.Pclass.value_counts().unstack().plot.bar(figsize=[10,10])
x2.set_title("Distribution of Survivors by Passenger Classes", fontsize=14)
x2.set_xlabel("Survived")
x2.set_ylabel("Passenger Count")
```

# Gender

```python
titanic_df.Sex.value_counts()
```

```python
x3 = titanic_df.Sex.value_counts().plot.bar()
x3.set_title("Distribution of Passengers by Gender", fontsize=14)
x3.set_xlabel("Gender")
x3.set_ylabel("Passenger Count")
```

```python
groups = titanic_df.groupby(by="Survived")
df_gender = groups.Sex.value_counts()
x4 = df_gender.unstack().plot.bar()
x4.set_title("Distribution of Survivors by Gender", fontsize=14)
x4.set_xlabel("Survived")
x4.set_ylabel("Passenger Count")
```

# Age

```python
titanic_df.Age.value_counts().head()
```

```python
x5 = titanic_df.Age.value_counts().plot.hist()
x5.set_title("Distribution of Passengers by Age", fontsize=14)
x5.set_xlabel("Age")
x5.set_ylabel("Passenger Count")
```

```python
groups_gender = titanic_df.groupby(by="Sex")
df_age_gender = groups_gender.Age.value_counts()
x10 = df_age_gender.unstack().plot.bar(legend = False)
x10.set_title("Distribution of Passengers by Age and Gender", fontsize=14)
x10.set_xlabel("Gender, Age")
x10.set_ylabel("Passenger Count")
```

```python
groups_gender.Age.mean()
```
