### Read in billboard.csv as a data frame

```python
df = pd.read_csv('../../assets/datasets/billboard.csv')
df
```

### The simplest pivot table must have a dataframe and an index. Let’s use the artist as our index.

```python
dfArtist = pd.pivot_table(df,index=["artist.inverted"])
dfArtist
```

### Set indices as "artist.inverted", "track" then use melt & stack, compare the differences

```python
df2 = df.copy()
df2.set_index(["artist.inverted", "track"], inplace=True)
df2.stack()
```

```python
mdf = pd.melt(df2.reset_index(), id_vars=df2.index.names)
mdf
```

```python
mdf.pivot_table(index=["artist.inverted","track"], columns='variable', values='value', aggfunc=min)
```

### Let's look at the data by artist and track. 

```python
dfArtistTrack = pd.pivot_table(df,index=["artist.inverted","track"])
dfArtistTrack
```

### Find out the number of weeks each song has been on the billboard

```python
# Find out what the x50th.week row actually means before solving this.

# This will show how many tracks are not null for each row
# dfArtistTrack.notnull().sum().sort_values()
# Shows one track
# dfArtistTrack.iloc[1].sort_values()

# Not to get complicated, remove year column
del dfArtistTrack['year']
```

```python
dfWeeksOnBillboard = dfArtistTrack.notnull().sum(axis=1)
dfWeeksOnBillboard
```

### Plot the number of weeks in billboard and tracks

```python
dfWeeksOnBillboard.plot(kind='hist', bins=50)
```

```python
# What is going on here?
sum(dfWeeksOnBillboard == 20.0)
```

```python
dfArtistTrackWeeks = pd.concat([dfWeeksOnBillboard, dfArtistTrack], axis=1)
dfArtistTrackWeeks
```
