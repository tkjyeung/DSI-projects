```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import sklearn.metrics as metrics
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = pd.DataFrame({'a':np.random.uniform(0.0, 100.0, size=1000)})
y = 50 + 3*x.a + np.random.normal(0.0, 30, size=1000)
print x.shape, y.shape
plt.scatter(x.a, y)
```

```python
lr = LinearRegression()
lr.fit(X=x, y=y)
print "y = {i} + {c}*x".format(i=lr.intercept_, c=lr.coef_[0])
```

```python
y_hat = lr.predict(X=x)
plt.scatter(y, y_hat)
print np.mean(y - y_hat)
print metrics.mean_squared_error(y, y_hat)
```

```python
x = pd.DataFrame({'a':np.random.uniform(0.0, 100.0, size=1000),'b':np.random.uniform(0.0, 100.0, size=1000)})
y = 50 + 3*x.a +4*x.b + np.random.normal(0.0, 30, size=1000)
print x.shape, y.shape
plt.scatter(x.a, y, c='r', marker='.')
plt.scatter(x.b, y, c='b', marker='.')
```

```python
lr = LinearRegression()
lr.fit(X=x, y=y)
print "y = {i} + {c}".format(i=lr.intercept_, c=" + ".join(["{}*x{}".format(coef, i) for i, coef in enumerate(lr.coef_)]))
```

```python
y_hat = lr.predict(X=x)
plt.scatter(y, y_hat)
print np.mean(y - y_hat)
print metrics.mean_squared_error(y, y_hat)
```
