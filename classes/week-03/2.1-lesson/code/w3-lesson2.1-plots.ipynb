# Bias Variance Tradeoff

In this notebook we investigate the bias and variance of models. In the first section we'll investigate fitting functions of the form:

$$f(x) = a + b * x^n$$

to some cubic data with noise.

## Note: This notebook generates the plots in the lesson readme and is neither the starter or solution code for the lesson.

```python
% matplotlib inline
import random

from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
from sklearn import datasets, linear_model
```

## Fit a linear model to quadratic data

```python
data = [(x, x * x + 1) for x in range(0, 20)]
df = pd.DataFrame(data, columns=["x", "y"])
```

```python
lm = linear_model.LinearRegression()

X = df[["x"]]
y = df["y"]

model = lm.fit(X, y)
predictions = lm.predict(X)

# Plot the model
plt.scatter(X, y, s=30, c='r', label="Raw Data")
plt.plot(X, predictions, label="Linear Fit")
# plt.xlabel()
# plt.ylabel("Actual Values MEDV")
plt.show()
```

```python
## Quadratic Model
```

```python
lm = linear_model.LinearRegression()

df["x^2"] = df["x"] * df["x"]

X = df[["x", "x^2"]]
y = df["y"]

model = lm.fit(X, y)
predictions = lm.predict(X)

# Plot the model
plt.scatter(df["x"], y, s=30, c='r', label="Raw Data")
plt.plot(df["x"], predictions, label="Linear Fit")
# plt.xlabel()
# plt.ylabel("Actual Values MEDV")
plt.show()
```
