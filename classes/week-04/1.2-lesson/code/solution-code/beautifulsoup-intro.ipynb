# Beautiful Soup

```python
from bs4 import BeautifulSoup
import urllib2
```

```python
url = "https://generalassemb.ly/education/data-science-immersive"

content = urllib2.urlopen(url).read()

soup = BeautifulSoup(content)

# Use function “prettify” to look at nested structure of HTML page
print soup.prettify()
```

## `soup.<tag>`
### Return content between opening and closing tag including tag.

```python
print soup.title
print soup.title.text
```

## Find all the links within page’s `<a>` tags

```python
for link in soup.find_all('a'):
    print(link.get('href'))
    print(link.text)
    
```
